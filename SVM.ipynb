# Support Vector Machine (SVM) 

A Support Vector Machine (SVM) is a supervised machine learning algorithm widely used for both linear and nonlinear classification, as well as regression and outlier detection tasks 

The primary objective of the SVM algorithm is to identify the optimal hyperplane in an N-dimensional space that can effectively separate data points into different classes in the feature space. The algorithm ensures that the margin between the closest points of different classes, known as support vectors, is maximized.

Hyperplane: A Hyperplane is simply the boundary or dividing line that separates the two groups (e.g., apples and oranges).
In 2D (on a flat surface), it’s just a line that separates the points into two groups.
In 3D, it’s a plane (like a sheet of paper) that separates the points.


Support Vectors:
Support Vectors are the data points that are closest to the boundary (the hyperplane) separating the two classes.
These are the critical points because the position of the boundary depends on them.

# Implementation 

```python
# Load the important packages
import matplotlib.pyplot as plt
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
import pandas as pd 
import numpy as np
```

```python
# Load the datasets
df = pd.read_csv('user_data.csv') 
x = df.iloc[:, [2, 3]].values  
y = df.iloc[:, 4].values 

#The features used to predict the target label (i.e., who will purchase and who will not) are:
#So, the input X will be the Age and EstimatedSalary columns.

#Output (y)
#Purchased: This column indicates whether the user purchased the product (1) or not (0).

#This shows the algorithm is supervised learning algorithm 
```

```python
df
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(x)  # Standardizing the features
```

```python
#Build the model
svm = SVC(kernel="rbf", gamma=0.5, C=1.0)

#kernel="rbf": Uses the Radial Basis Function kernel to transform the data for non-linear separation.
#gamma=0.5: Controls how much influence each data point has; a moderate value helps balance the decision boundary.
#C=1.0: Regularization parameter that balances fitting the training data and keeping the model simple to avoid overfitting.
```

```python
# Trained the model
svm.fit(x, y)
```

```python
# Plot Decision Boundary
disp = DecisionBoundaryDisplay.from_estimator(
    svm, x, response_method="predict", cmap=plt.cm.Spectral, alpha=0.8
)

#cmap defines the color map to use for visualizing different classes in the decision boundary plot.
#plt.cm.Spectral is a specific color map that assigns different colors to different classes in the plot (from red to purple, typically).
#alpha=0.8-alpha controls the transparency of the plot. A value of 0.8 means the plot will be 80% opaque

plt.scatter(x[:, 0], x[:, 1], c=y, s=20, edgecolors="k", cmap=plt.cm.coolwarm)
plt.xlabel(df.columns[2])  # Use feature names
plt.ylabel(df.columns[3])
plt.title("SVM Decision Boundary")
plt.show()

# c=y:c specifies the color of the points in the scatter plot.
# y represents the class labels of the data points, so each point will be colored according to its class label. Typically, this will be an array of integers or categories that represent different classes.
#s=20:s specifies the size of the points (markers) in the scatter plot.
#edgecolors="k":edgecolors defines the color of the edges of the points.
```

 the goal of this SVM model is to learn a boundary (decision boundary) that can separate the users who will purchase the product (Purchased = 1) from those who won’t (Purchased = 0), based on their age and estimated salary.
