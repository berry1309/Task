# Linear Regression 


Linear regression is a method used in machine learning that learns from labelled datasets and helps us to understand relationship between a dependent variable and one or more independent variables.

It is a mathematical model that assumes a linear relationship between the input (dependent) variables (X) and the output variable( Independent) (Y). It tries to fit a straight line through a set of data points in a way that minimizes the differences between the actual and predicted values. It performs the task to predict a dependent variable value (output) based on a given independent variable (input)). 

Equation: 
Y = mX + C 

the intercept represents the point where the line crosses the Y-axis in a simple linear regression 
The slope indicates how much the target variable (Y) changes when a specific input variable (X) changes. 
For example, if m is 2, it means for every 1-unit increase in X, Y will increase by 2 units.

The main task is to find the best fit line, which indicates that the error between the predicted and actual values should be kept to a minimum. Naturally, best fit line will have the least error.
The slope of the line indicates how much the dependent variable changes for a unit change in the independent variables.

# Algorithm 

Step 1: Create a formula to relate the input (X) to the output (Y).
    This formula will have some unknown values (coefficients) that we need to figure out.
    
Step 2: Find the values for these unknowns that make the formula as accurate as possible in predicting the output (Y).
    
Step 3: Calculate how far off your predictions are from the actual results. We want to minimize this error.
    
Step 4: After training the model, use the formula to predict new results based on new inputs.

# Code

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Taking sample data where,
# X is the innput feature 
# Y is the output 
X = np.array([[1], [2], [3], [4], [5]]) 
Y = np.array([1, 2, 2.5, 4.5, 5]) 
```

```python
model = LinearRegression()
LinearRegression()
```

```python
#the library sci-kit learn has inbuilt linear regression model 
```

```python
model.fit(X, Y)
```



#After the model has been trained (i.e., it has the values for the slope and intercept), it can use the equation Y=mX+b to make predictions on new data.

For instance, if X = 6 (6 years of experience), the model will predict Y (the salary) as:

Y=1.06+0.5=6.5

So, the model predicts a salary of 6.5 for someone with 6 years of experience.

```python
Y_pred = model.predict(X)
```

```python
print("Intercept:", model.intercept_)
print("Slope:", model.coef_)
#If the slope is negative, then an increase in X will result in a decrease in Y.
```

```python
plt.scatter(X, Y, color='purple')  # Actual data points
plt.plot(X, Y_pred, color='pink')  # Predicted regression line
plt.xlabel('Input Feature (X)')
plt.ylabel('Output (Y)')
plt.title('Linear Regression')
plt.show()
```
